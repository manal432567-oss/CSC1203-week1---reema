```python
age = int(input("Enter your age: "))

if age < 13:
    print("Child")
elif age <= 19:
    print("Teen")
elif age <= 59:
    print("Adult")
else:
    print("Senior")

```

```python
num = float(input("Enter a number: "))

if num > 0:
    print("Positive")
else:
    print("Not Positive")

if 10 <= num <= 50:
    print("Number is between 10 and 50")
```

```python
color = input("Enter a color: ")

if color != "":
    match color.lower():
        case "red":
            print("Stop")
        case "yellow":
            print("Ready")
        case "green":
            print("Go")
        case _:
            print("Unknown color")
else:
    print("Input cannot be empty")
```

```python
print("1) Addition")
print("2) Subtraction")
print("3) Multiplication")
print("4) Division")

choice = input("Choose option: ")

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

match choice:
    case "1":
        print("Result =", num1 + num2)
    case "2":
        print("Result =", num1 - num2)
    case "3":
        print("Result =", num1 * num2)
    case "4":
        if num2 != 0:
            print("Result =", num1 / num2)
        else:
            print("Cannot divide by zero")
    case _:
        print("Invalid choice")
```

```python
score = float(input("Enter student score: "))

if score >= 50:
    print("Passed")
else:
    print("Failed")

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

match grade:
    case "A":
        print("Excellent")
    case "B":
        print("Very Good")
    case "C":
        print("Good")
    case "D":
        print("Pass")
    case "F":
        print("Try again")
```
