```python

total = 0

for i in range(10):
    num = int(input("Enter number: "))

    if num == 0:
        break

    if num < 0:
        continue

    if num % 2 == 0:
        total += num

print("Final sum =", total)
```

```python
n = int(input("Enter n: "))
count = 0

for i in range(1, n + 1):
    if i > 50:
        break

    if i % 8 == 0:
        continue

    if i % 4 == 0:
        count += 1

print("Count =", count)
```

```python
secret = 7
attempts = 0

while True:
    guess = int(input("Guess number (1-20): "))

    if guess < 1 or guess > 20:
        continue

    attempts += 1

    if guess == secret:
        print("Correct! Attempts =", attempts)
        break
```

```python
n = int(input("Enter a number: "))

for i in range(1, 13):
    result = n * i

    if result % 3 == 0:
        continue

    if result > 50:
        break

    print(n, "x", i, "=", result)
```

```python
for i in range(1, 6):
    for j in range(1, i + 1):
        if j == 3:
            continue

        if i * j > 6:
            break

        print(j, end=" ")
    print()
```
